```python
import random
from PIL import Image
from pathlib import Path
import os

from python_file.dirPath import csvDir, txtDir, imageDir, dayTest, fogTest, nightTest

inizioFile = 3000
```

```python
# Leggi le annotazioni
with open(dayTest / 'prova.txt') as f:
    annotationsDay = [line.strip().split() for line in f.readlines()]
    
# Leggi le annotazioni
with open(fogTest / 'prova.txt') as f:
    annotationsFog = [line.strip().split() for line in f.readlines()]
    
# Leggi le annotazioni
with open(nightTest / 'prova.txt') as f:
    annotationsNight = [line.strip().split() for line in f.readlines()]
```

```python
# Raggruppa le annotazioni per immagine
from collections import defaultdict
ann_per_img_day = defaultdict(list)
for ann in annotationsDay:
    ann_per_img_day[ann[0]].append([int(ann[1]), int(ann[2]), int(ann[3]), int(ann[4])])
    
ann_per_img_fog = defaultdict(list)
for ann in annotationsFog:
    ann_per_img_fog[ann[0]].append([int(ann[1]), int(ann[2]), int(ann[3]), int(ann[4])])
    
ann_per_img_night = defaultdict(list)
for ann in annotationsNight:
    ann_per_img_night[ann[0]].append([int(ann[1]), int(ann[2]), int(ann[3]), int(ann[4])])

print(len(ann_per_img_day))  
print(len(ann_per_img_fog))  
print(len(ann_per_img_night))  
```

```python
bg_count = inizioFile

numImage = 80

file = open(txtDir / "class4.txt", "w")

for i in range(numImage):

    ann_per_img = defaultdict(list)
    
    scelta   = random.randint(1,3)
    if scelta == 1:
        immagine = random.randint(0, len(ann_per_img_day))-1
        #print("Day")
        #print(immagine)
        img_name = list(ann_per_img_day.keys())[immagine]
        boxes = ann_per_img_day[img_name]
        #print(img_name)
        baseDir = dayTest / img_name
    elif scelta == 2:
        immagine = random.randint(0, len(ann_per_img_fog)-1)
        #print("Fog")
        #print(immagine)
        img_name = list(ann_per_img_fog.keys())[immagine]
        boxes = ann_per_img_fog[img_name]
        #print(img_name)
        baseDir = fogTest / img_name
    elif scelta == 3:
        immagine = random.randint(0, len(ann_per_img_night)-1)
        #print("Night")
        #print(immagine)
        img_name = list(ann_per_img_night.keys())[immagine]
        boxes = ann_per_img_night[img_name]
        #print(img_name)
        baseDir = nightTest / img_name
    else:
        continue

    img = Image.open(baseDir)
    w, h = img.size
    
    for _ in range(5):  # 5 ritagli background per immagine
        size = random.choice([64, 96, 128, 160])
        if size >= w or size >= h:
            continue
        x = random.randint(0, w - size)
        y = random.randint(0, h - size)
        
        # Verifica che NON sovrapponga nessun segnale
        is_bg = True
        for (bx1, by1, bx2, by2) in boxes:
            if not (x + size < bx1 or x > bx2 or y + size < by1 or y > by2):
                is_bg = False
                break
        
        if is_bg:
            crop = img.crop((x, y, x + size, y + size))
            crop.save(imageDir / f"{bg_count}.png")
            contenuto = f"{bg_count}.png, 3\n"
            file.write(contenuto)
            bg_count += 1
            
file.close()
print(f"Generati {bg_count-inizioFile} campioni background")
```
